```python
# division in the buckets of low i.e. below 10000, medium i.e. range b/w 10000 and 20000 and high i.e. above 20000

category = [
    (CarData['price'] < 10000),
    (CarData['price'] >= 10000) & (CarData['price'] <= 20000),
    (CarData['price'] > 20000)
]
labels = ['low', 'medium', 'high']
CarData['price_category'] = np.select(category, labels, default='unknown')

# mapping the company_price on the 'CarCompany' column in the dataset
CarData['CarCompany_Median_Price'] = CarData['CarCompany'].map(company_price_median)
# CarData.head()

# checking the column's unique values
CarData.CarCompany.unique()
```

```python
#creating dummy variable for catwegorical variables

categorical_cols = ['fueltype', 'aspiration', 'doornumber', 'carbody', 'enginelocation', 'cylindernumber', 'drivewheel', 'enginetype']

CarDataWDummies = pd.get_dummies(CarData, columns=categorical_cols, drop_first=True)

#checking the dataset
CarDataWDummies.head()
```

```python
# importing required library
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively

CarData_train, CarData_test = train_test_split(CarDataWDummies, test_size=0.3, random_state=42)

# shape of the train and test sets
print(CarData_train.shape)
print(CarData_test.shape)
# (143, 203)
# (62, 203)
```

```python
#importing required function for scaling

from sklearn.preprocessing import MinMaxScaler

columns_to_scale = [
    'wheelbase', 'carlength', 'carwidth', 'carheight',
    'curbweight', 'enginesize', 'boreratio', 'stroke',
    'compressionratio', 'horsepower', 'peakrpm', 'citympg',
    'highwaympg', 'CarCompany_Median_Price'
]
scaler = MinMaxScaler()

#Scale only the specified columns
scaled_train_values = scaler.fit_transform(CarData_train[columns_to_scale])
scaled_test_values = scaler.transform(CarData_test[columns_to_scale])

# Create new DataFrames with scaled values for specified columns and original values for others
CarData_train_scaled = CarData_train.copy()
CarData_train_scaled[columns_to_scale] = scaled_train_values

CarData_test_scaled = CarData_test.copy()
CarData_test_scaled[columns_to_scale] = scaled_test_values

# Verify the scaling
print("Original training data range:")
print(CarData_train[columns_to_scale].describe().loc[['min', 'max']])

print("\nScaled training data range:")
print(CarData_train_scaled[columns_to_scale].describe().loc[['min', 'max']])

```

```python
# Get all numerical columns except price
all_numerical = CarData_train_scaled.select_dtypes(include=['int64', 'float64']).columns.tolist()
feature_columns = [col for col in all_numerical if col != 'price']

# Remove any other columns you don't want (like IDs, indexes)
columns_to_exclude = ['car_ID', 'Unnamed: 0']  # Add any other columns to exclude
feature_columns = [col for col in feature_columns if col not in columns_to_exclude]

X_train = CarData_train_scaled[feature_columns]
X_test = CarData_test_scaled[feature_columns]

y_train = CarData_train_scaled['price']
y_test = CarData_test_scaled['price']

# print(f"Final feature set: {len(feature_columns)} columns")

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
# Train your model
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# Fit the model on the scaled training data
model.fit(X_train, y_train)

# Check the model coefficients
print("Model intercept:", model.intercept_)
print("Model coefficients:", model.coef_)
print("Number of coefficients:", len(model.coef_))

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
from sklearn.metrics import mean_squared_error, r2_score

print("\nModel Evaluation:")
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R-squared Score:", r2_score(y_test, y_pred))
print("So, our model explain ", r2_score(y_test, y_pred), "% of the variance in car prices")
```

```python
# -*- coding: utf-8 -*-
"""Beginner02 - Implementao manual sem bibliotecas externas"""

import csv
import math
import random

# Funo para ler o arquivo CSV
def read_csv(filename):
    data = []
    with open(filename, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            data.append(row)
    return data

# Funo para calcular estatsticas descritivas
def calculate_descriptive_stats(data, column):
    values = [float(row[column]) for row in data if row[column].strip()]
    n = len(values)

    if n == 0:
        return None

    mean = sum(values) / n
    sorted_values = sorted(values)

    # Mediana
    if n % 2 == 0:
        median = (sorted_values[n//2 - 1] + sorted_values[n//2]) / 2
    else:
        median = sorted_values[n//2]

    # Desvio padro
    variance = sum((x - mean) ** 2 for x in values) / (n - 1)
    std_dev = math.sqrt(variance)

    return {
        'count': n,
        'mean': mean,
        'std_dev': std_dev,
        'min': min(values),
        'max': max(values),
        'median': median
    }

# Funo para extrair a empresa do nome do carro
def extract_car_company(car_name):
    return car_name.split(' ')[0].lower()

# Corrigir nomes de empresas com erros de digitao
def correct_company_names(data):
    replacements = {
        'maxda': 'mazda',
        'nissan': 'nissan',
        'porcshce': 'porsche',
        'toyouta': 'toyota',
        'vokswagen': 'volkswagen',
        'vw': 'volkswagen'
    }

    for row in data:
        company = extract_car_company(row['CarName'])
        row['CarCompany'] = replacements.get(company, company)

# Funo para calcular R score manualmente
def r2_score(y_true, y_pred):
    y_true = [float(y) for y in y_true]
    y_pred = [float(y) for y in y_pred]

    # Valor mdio dos valores reais
    y_mean = sum(y_true) / len(y_true)

    # Soma total dos quadrados
    ss_total = sum((y - y_mean) ** 2 for y in y_true)

    # Soma dos quadrados dos resduos
    ss_residual = sum((y_true[i] - y_pred[i]) ** 2 for i in range(len(y_true)))

    # R score
    r2 = 1 - (ss_residual / ss_total)
    return r2

# Funo para calcular MSE manualmente
def mean_squared_error(y_true, y_pred):
    y_true = [float(y) for y in y_true]
    y_pred = [float(y) for y in y_pred]

    mse = sum((y_true[i] - y_pred[i]) ** 2 for i in range(len(y_true))) / len(y_true)
    return mse

# Funo para normalizao Min-Max manual
def min_max_scale(data, columns):
    scaled_data = data.copy()

    for col in columns:
        values = [float(row[col]) for row in data if row[col].strip()]
        min_val = min(values)
        max_val = max(values)

        for row in scaled_data:
            if row[col].strip():
                original_val = float(row[col])
                scaled_val = (original_val - min_val) / (max_val - min_val)
                row[col] = scaled_val

    return scaled_data

# Funo para dividir dados em treino e teste
def train_test_split(data, test_size=0.3, random_state=None):
    if random_state is not None:
        random.seed(random_state)

    random.shuffle(data)
    split_index = int(len(data) * (1 - test_size))

    train_data = data[:split_index]
    test_data = data[split_index:]

    return train_data, test_data

# Implementao manual de regresso linear
class ManualLinearRegression:
    def __init__(self):
        self.coefficients = []
        self.intercept = 0

    def fit(self, X, y):
        # Adicionar coluna de 1s para o intercept
        X_with_intercept = [[1.0] + list(x) for x in X]

        # Converter para matriz (lista de listas)
        n = len(X_with_intercept)
        m = len(X_with_intercept[0])

        # Mtodo dos mnimos quadrados: (X'X)^-1 X'y
        # Calcular X'X
        xtx = [[0.0] * m for _ in range(m)]
        for i in range(m):
            for j in range(m):
                xtx[i][j] = sum(X_with_intercept[k][i] * X_with_intercept[k][j] for k in range(n))

        # Calcular X'y
        xty = [0.0] * m
        for i in range(m):
            xty[i] = sum(X_with_intercept[k][i] * y[k] for k in range(n))

        # Resolver sistema (simplificado - para pequenos datasets)
        # Para datasets maiores, precisaria de uma implementao mais robusta
        self.coefficients = self.solve_system(xtx, xty)
        self.intercept = self.coefficients[0]
        self.coefficients = self.coefficients[1:]

    def solve_system(self, A, b):
        # Implementao simplificada de eliminao gaussiana
        n = len(b)

        # Eliminao direta
        for i in range(n):
            # Pivotizao parcial
            max_row = i
            for k in range(i + 1, n):
                if abs(A[k][i]) > abs(A[max_row][i]):
                    max_row = k

            # Trocar linhas
            A[i], A[max_row] = A[max_row], A[i]
            b[i], b[max_row] = b[max_row], b[i]

            # Eliminar
            for k in range(i + 1, n):
                factor = A[k][i] / A[i][i]
                for j in range(i, n):
                    A[k][j] -= factor * A[i][j]
                b[k] -= factor * b[i]

        # Substituio reversa
        x = [0.0] * n
        for i in range(n - 1, -1, -1):
            x[i] = b[i]
            for j in range(i + 1, n):
                x[i] -= A[i][j] * x[j]
            x[i] /= A[i][i]

        return x

    def predict(self, X):
        predictions = []
        for x in X:
            prediction = self.intercept
            for i in range(len(self.coefficients)):
                prediction += self.coefficients[i] * x[i]
            predictions.append(prediction)
        return predictions

# Funo principal
def main():
    # Ler dados
    print("Lendo dados do arquivo CSV...")
    car_data = read_csv('Car_Price.csv')

    # Informaes bsicas
    print(f"Total de registros: {len(car_data)}")
    print(f"Total de colunas: {len(car_data[0].keys())}")

    # Extrair empresa do carro
    print("\nExtraindo empresa dos carros...")
    for row in car_data:
        row['CarCompany'] = extract_car_company(row['CarName'])

    # Corrigir nomes
    correct_company_names(car_data)

    # Calcular preo mdio por empresa
    company_prices = {}
    for row in car_data:
        company = row['CarCompany']
        price = float(row['price'])

        if company not in company_prices:
            company_prices[company] = []
        company_prices[company].append(price)

    company_median_prices = {}
    for company, prices in company_prices.items():
        sorted_prices = sorted(prices)
        n = len(sorted_prices)
        if n % 2 == 0:
            median = (sorted_prices[n//2 - 1] + sorted_prices[n//2]) / 2
        else:
            median = sorted_prices[n//2]
        company_median_prices[company] = median

    # Adicionar preo mediano ao dataset
    for row in car_data:
        row['CarCompany_Median_Price'] = company_median_prices[row['CarCompany']]

    # Selecionar features numricas para o modelo
    numeric_features = [
        'wheelbase', 'carlength', 'carwidth', 'carheight',
        'curbweight', 'enginesize', 'boreratio', 'stroke',
        'compressionratio', 'horsepower', 'peakrpm', 'citympg',
        'highwaympg', 'CarCompany_Median_Price'
    ]

    # Preparar dados para o modelo
    X = []
    y = []

    for row in car_data:
        try:
            # Extrair features numricas
            features = [float(row[feature]) for feature in numeric_features]
            target = float(row['price'])

            X.append(features)
            y.append(target)
        except (ValueError, KeyError):
            # Pular linhas com dados invlidos
            continue

    # Dividir em treino e teste
    print("\nDividindo dados em treino e teste...")
    train_X, test_X = train_test_split(X, test_size=0.3, random_state=42)
    train_y, test_y = train_test_split(y, test_size=0.3, random_state=42)

    print(f"Dados de treino: {len(train_X)} amostras")
    print(f"Dados de teste: {len(test_X)} amostras")

    # Treinar modelo
    print("\nTreinando modelo de regresso linear...")
    model = ManualLinearRegression()
    model.fit(train_X, train_y)

    # Fazer previses
    print("Fazendo previses...")
    predictions = model.predict(test_X)

    # Avaliar modelo
    print("\nAvaliando modelo:")
    mse = mean_squared_error(test_y, predictions)
    r2 = r2_score(test_y, predictions)

    print(f"Mean Squared Error: {mse:.2f}")
    print(f"R Score: {r2:.4f}")
    print(f"O modelo explica {r2 * 100:.2f}% da varincia nos preos dos carros")

    # Mostrar algumas previses vs valores reais
    print("\nAlgumas previses vs valores reais:")
    for i in range(min(5, len(test_y))):
        print(f"Real: {test_y[i]:.2f}, Predito: {predictions[i]:.2f}")

if __name__ == "__main__":
    main()
```
